```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

x, y = np.meshgrid(
    np.array([(i + 0.5) / 50 for i in range(-15, 15)]),
    np.array([(i + 0.5) / 50 for i in range(-15, 15)])
)

z = x + 1j * y

rfz = np.sin((1/z).real)
ifz = np.sin((1/z).imag)
```

```python
fig = plt.figure(facecolor='black')

axr = fig.add_subplot(1,2,1, projection='3d', facecolor='black')
axi = fig.add_subplot(1,2,2, projection='3d', facecolor='black')

def update(step):
    axr.clear()
    axr.grid(color='grey')
    axr.view_init(elev=step / 10 % 90, azim=step % 360)
    axr.plot_wireframe(x, y, rfz, color='blue', linewidth=1)
    axr.set_xlabel('$x$', color='white')
    axr.set_ylabel('$y\cdot i$', color='white')
    axi.tick_params(axis='x', color='white')
    axi.tick_params(axis='y', color='white')
    axr.set_title('$f(z)=\\sin {\\rm Re}(1/z)$', color='white')

    axi.clear()
    axi.grid(color='grey')
    axi.view_init(elev=step / 10 % 90, azim=step % 360)
    axi.plot_wireframe(x, y, ifz, color='red', linewidth=1)
    axi.set_xlabel('$x$', color='white')
    axi.set_ylabel('$y\cdot i$', color='white')
    axi.tick_params(axis='x', color='white')
    axi.tick_params(axis='y', color='white')
    axi.set_title('$f(z)=\\sin {\\rm Im}(1/z)$', color='white')

    return plt

anim = FuncAnimation(fig=fig, func=update, frames=720, interval=10)
anim.save('complex-plane-singularity-1.gif', dpi=200)
```
