```python
ptx = -10
pty = -10

def f(x, y): return (x**2 - y**2 + 1 * x * y) / 100
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

fig = plt.figure(figsize=(30, 10))
ax1 = fig.add_subplot(1, 3, 1, projection='3d', facecolor='black')
ax2 = fig.add_subplot(1, 3, 2, facecolor='black')
ax3 = fig.add_subplot(1, 3, 3, facecolor='black')

η = 5e-1
sgn = 1

def update(step):
    global ptx, pty, sgn

    for ax in [ax1, ax2, ax3]:
        ax.clear()
        ax.xaxis.label.set_color('white')
        ax.yaxis.label.set_color('white')
        ax.tick_params(axis='x', colors='white')
        ax.tick_params(axis='y', colors='white')
        ax.spines['left'].set_color('white')
        ax.spines['right'].set_color('white')
        ax.spines['bottom'].set_color('white')
        ax.spines['top'].set_color('white')

    ax1.zaxis.label.set_color('white')
    ax1.tick_params(axis='z', colors='white')

    ax2.set_xlim(-10, 10)
    ax2.set_ylim(-5, 5)
    ax3.set_xlim(-10, 10)
    ax3.set_ylim(-5, 5)

    X, Y = np.meshgrid(np.linspace(-10, 10, 50), np.linspace(-10, 10, 50))

    ax1.plot_wireframe(X, Y, f(X, Y), rstride=0, cstride=2,
                       color='red', alpha=0.7, linewidth=1)
    ax1.plot_wireframe(X, Y, f(X, Y), rstride=2, cstride=0,
                       color='blue', alpha=0.7, linewidth=1)
    ax1.scatter3D([ptx], [pty], [f(ptx, pty)], marker='x', color='white', s=80)
    ax1.set_title(r'$f(x)=\frac{1}{100}(x^2+y^2+x\cdot y)$', color='white')

    ax2.plot(np.linspace(-10, 10, 50),
             f(np.linspace(-10, 10, 50), pty), color='blue')
    ax2.scatter([ptx], [f(ptx, pty)], marker='x', color='white', s=80)
    ax2.set_title('x', color='white')

    ax3.plot(np.linspace(-10, 10, 50),
             f(ptx, np.linspace(-10, 10, 50)), color='red')
    ax3.scatter([pty], [f(ptx, pty)], marker='x', color='white', s=80)
    ax3.set_title('y', color='white')
    ptx = ptx + (η / 20)

    if pty >= 10:
        sgn = -1
    elif pty <= -10:
        sgn = 1

    pty = pty + sgn*η
    # 20 / η steps for traversing

    return fig

FuncAnimation(
    fig=fig,
    func=update,
    frames=int(20*20 / η),
    # frames=20,
    interval=int(50 * η)
).save('../Images/min-max-theorem-illustrated.gif', dpi=100)
```
